```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import sklearn
from sklearn.datasets import make_blobs

from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale

import sklearn.metrics as sm
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import MinMaxScaler
```

```python
irisData = pd.read_csv('army.csv')
irisData
```

```python

plt.rcParams["figure.figsize"] = (15,15)
plt.scatter(irisData['2017'],irisData['Country Name'])
```

```python
km = KMeans(n_clusters = 3)
y_predicted = km.fit_predict(irisData[['2017','ID']])
```

```python
irisData['cluster'] = y_predicted
```

```python
km.cluster_centers_

df1 = irisData[irisData.cluster == 0]
df2 = irisData[irisData.cluster == 1]
df3 = irisData[irisData.cluster == 2]

plt.scatter(df1['2017'],df1['ID'], color = 'yellow')
plt.scatter(df2['2017'],df2['ID'], color = 'red')
plt.scatter(df3['2017'],df3['ID'], color = 'skyblue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], color = 'black', marker = '*', label = 'centroid')
```
